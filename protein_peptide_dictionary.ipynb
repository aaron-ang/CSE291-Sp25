```python
import pandas as pd

file_path = 'data/mq_variants_intensity.tsv'
data = pd.read_csv(file_path, sep='\t', header=None)

protein_column_index = 8 
peptide_column_index = 2  

protein_to_peptides = {}

for _, row in data.iloc[1:].iterrows():
    protein_entry = row[protein_column_index]
    peptide = row[peptide_column_index]
    
    if pd.notna(protein_entry) and pd.notna(peptide):
        proteins = protein_entry.split(';')
        if len(proteins) == 1:
            protein = proteins[0]
            if '|' in protein:
                parts = protein.split('|')
                if len(parts) > 2:
                    protein_key = parts[1]
                    if protein_key not in protein_to_peptides:
                        protein_to_peptides[protein_key] = set()
                    protein_to_peptides[protein_key].add(peptide)

print(protein_to_peptides)
print(len(protein_to_peptides))
```

```python

for i, (protein, peptides) in enumerate(protein_to_peptides.items()):
    print(f"{protein}: {list(peptides)[:5]}")
    if i >= 2: 
        break
```

```python
for protein, peptides in protein_to_peptides.items():
    print(f"Protein: {protein}, Number of Peptides: {len(peptides)}")
```

```python
with open('data/protein_to_peptides.tsv', 'w') as f:
    f.write("Protein\tPeptides\n")
    for protein, peptides in protein_to_peptides.items():
        peptides_str = ";".join(peptides)  # Join peptides with a semicolon
        f.write(f"{protein}\t{peptides_str}\n")
```

```python

filtered_proteins = {protein: peptides for protein, peptides in protein_to_peptides.items() if len(peptides) >= 10}

with open('data/protein_with_10+_peptides.tsv', 'w') as f:
    f.write("Protein\tPeptide_Count\tPeptides\n")
    for protein, peptides in filtered_proteins.items():
        peptides_str = ";".join(peptides)  # Join peptides with a semicolon
        f.write(f"{protein}\t{len(peptides)}\t{peptides_str}\n")
```
